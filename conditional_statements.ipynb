```python
# Conditional Statements
# 1.if
# 2.elif
# 3.elss
# Conditional Statements are used to control the flow of execution in a program based on specific condition.
```

```python
# 1. if statement
age = int(input("enter Your age".title()))
if age >= 18:
    print("Eligible for vote")
          
```

```python
# 2. if-else statement
age = int(input("Enter Your age".title()))
if age >= 18:
    print("eligible for voting".title())
else:
    print("not eligible for voting".title())
```

```python
# Check even or odd
number = int(input("Enter your number: "))
if number % 2 == 0:
    print("Even")
else:
    print("odd")
```

```python
# 3.if-elif-else statements
marks = int(input("Enter Your marks: "))
if marks >= 80:
    print("A Grade")
elif marks >= 65:
    print("B Grade")
else:
    print("C Grade")
    
```

```python
# Nested if statements
age = int(input("Enter your age:"))
citizen = True
if age >= 18:
    if citizen:
        print("Eligible for voting")
```

```python
# Multiple conditions (and)
# if we using and both conditions must be True.
age = 25
salary = 50000
if age >= 25 and salary >= 30000:
    print("Loan Approved")
```

```python
#multiple conditions (or)
# At least one condition must be True.
day = "Sunday"
if day == "Saturday" or day == "Sunday":
    print("Weekend")
    
```

```python
fruits = ['Apple', 'Banana', 'Orange']
if 'Apple' in fruits:
    print("Found")
```

```python
# Match-Case
day = 2
match day:
    case 1:
        print('monday')
    case 2:
        print ('Tuesday')
    case 3:
        print('Wednesday')
    case _:
        print("Invalid!")
```

```python
# 1. Check Positive, Negative, or zero
number = int(input("Enter Your Number: "))
if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Find the largest Two Number
number_1 = int(input("Enter Your first Number: "))
number_2 = int(input("Enter Your Second Number: "))
if number_1 > number_2:
    print(f"The Largest Number is {number_1}")
elif number_2 > number_1:
    print(f"The largest Number is {number_2}")
else:
    print("Both Number are same")
```

```python
# Find the largest Three Number
num1 = int(input("Enter Your First Number: "))
num2 = int(input("Enter Your second number: "))
num3 = int(input("Enter Your Third number: "))
if num1 > num2 :
    print(f"The Largest Number is:{num1}".title())
elif num2 > num3:
    print(f"The Largest number is : {num2}".title())
elif num3 > num1:
    print(f"The Largest number is : {num3}".title())
else:
    print("Three number are same")
```

```python
# Calculator 
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
operators = input("Enter operators (+,-,*,/,%): ")
if operators == "+":
    print(a+b)
elif operators == "-":
    print(a-b)
elif operators == "*":
    print(a*b)
elif operators == "/":
    if b != 0:
        print(a/b)
    else:
        print("cannot divided by zero")
elif operators == "%":
    print(a%b)
else: 
    print("Invalid syntax")
```

```python
# ATM Machine
balance = 1000
pin = int(input("Enter Your pin: "))
if pin == 1234:
    amount = int(input("Withdraw Amount: "))
    if amount <= balance:
        balance -= amount
        print("Withdraw successful")
        print("Remaing Balance: ",balance)
    else:
        print("Insufficient Balance")
else:
    print("Invalid Pin")
    
```
