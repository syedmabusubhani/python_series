```python
# Student Management System
class Student:
    school = "Dreamz highschool"
    def __init__(self,name,age,marks):
        self.name = name
        self.age = age
        self.marks = marks
    def display(self):
        print(f"name : {self.name}")
        print(f"Age : {self.age}")
        print(f"marks : {self.marks}")
        print(f"school : {Student.school}")
    def grade(self):
        if self.marks >= 90:
            print("Grade A")
        elif self.marks >= 75:
            print("Grade B")
        elif self.marks >= 60:
            print("Grade C")
        else: 
            print("Fail")
s1 = Student("Rahul",21,92)
s1.display()
s1.grade()
```

```python
# Employee Class With Salary
class Employee:
    def __init__(self,emp_name,salary):
        self.emp_name = emp_name
        self.salary = salary
    def display(self):
        print(f"Employee Name : {self.emp_name}")
        print(f"salary        : {self.salary}")
e1 = Employee("Sony", 50000)
e2 = Employee("Sana", 70000)
e3 = Employee("Yafiya", 40000)
e4 = Employee("Alia", 30000)
e5 = Employee("Fizzy", 45990)
e1.display()
print("---------------------")
e2.display()
print("---------------------")
e3.display()
print("---------------------")
e4.display()
print("---------------------")
e5.display()
```

```python
# Bank Account Class
class BankAccount:
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance
    def deposit(self,amount):
        self.balance += amount
        print(f"Deposited : {amount}")
    def withdraw(self,amount):
        if amount <= self.balance
            self.balance -= amount
            print("Withdraw:",amount)
        else:
            print("Insufficient balance")
    def display(self):
        print("Account Holder:",self.name)
        print("Current Balance:",self.balance)
acc = BankAccount("sony", 10000)
acc.display()
acc.deposit(5000)
acc.withdraw(2000)
acc.display()
```
