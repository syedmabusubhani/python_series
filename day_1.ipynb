```python
# Variables and Data types
# variables: variables are used to store the data
x = 5 # here x is a variable
name = 'Alice' # here name is a variable
print(x)
print(name)
```

```python
# Assigning values to variables
x = 2
y = 1.9
z = 'HI!'
print(x,y,z)
# Assigning same values
a = b = c = 100
print(a,b,c)
# Assigning different values
a, b, c = 1,1.2,"python"
print(a,b,c)
# deleting a variable
x = 10
print(x)
del x
```

```python
# Data types
a = 2  # integer  data type
print(type(a))
b = 3.14 # floating data type
print(type(b))
c = 'Python Program' # string data type
print(type(c))
d = "Hello world!" # strings
print(type(d))
e = True # boolean
print(type(e))
f = 3+4j # complex
print(type(f))
```

```python
# Working with string
# string is a collection of charachters inside in double quotes (" "), single quotes (' ').
# Creating a string
name = "Alex"
city = 'Bengalore'
message = """ Welcome to Python"""
print (f"{name}\n{city}\n{message}")
```

```python
# Accessing Charactes (indexing)
text = "PYTHON"
print(text[0])
print(text[1])
print(text[5])
print(text[-1])
print(text[-4])
```

```python
# string Slicing
# string[start:end:step]
greats = "Welcome to Python"
print(greats)
print(greats[0:7])
print(greats[11:17])
print(greats[2:])
print(greats[:4])
print(greats[::-1])
print(greats[0:7:1])
print(greats[0:7:2])
```

```python
# string Length
string = "I love india"
print(len(string))
```

```python
# String Concatenation
first_str = "Data"
second_str = "Science"
print(first_str+ " "+second_str)
con_str = first_str+" "+second_str
print(con_str)
```

```python
# string Repetition
print("Hi " * 3)
```

```python
# Membership Operations
text = "Python Programming"
print("Python" in text)
print("java" in text)
```

```python
# Escape Characters
print("Hello\nWorld")
print("hello\tWorld") 
```

```python
# String Methods
# 1.Upper()
str = "python"
print(str.upper())
# 2. Lower()
str1 = "PYTHON"
print(str1.lower())
# 3. title
str2 = "python programming"
print(str2.title())
# 4. count()
print(str2.count("p"))
```

```python
# Examples
# Palindrome Check
word = "madam"
if word == word[::-1]:
    print("palindrome".title())
else:
    print("not palindrome".title())
```

```python
# Count vowels 
text = "Python Programming"
count = 0
for i in text.lower():
    if i in "aeiou":
        count += 1
print(f"count is :{count}")
```

```python
# Email Validation
user_email = input("Enter Your Email")
if "@" in user_email and "." in user_email:
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
# Password Strength
user_password = input("Enter your password ")
if len(user_password) >= 5:
    print("strong password".title())
else:
    print("weak password".title())
```
